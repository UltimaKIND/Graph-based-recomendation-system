```python
# 5 популярных жанров
popular_genres = data[['genre', 'track_id']].groupby('genre').count()\
.sort_values('track_id', ascending=False).head(5)\
.reset_index().drop('track_id', axis=1)
```

```python
popular_genres
```

```python
#Треки из 5 популярных жанров музыки
popular_tracks = data[['track_id', 'genre', 'name', 'artist']][data['genre'].isin(popular_genres['genre'])]
```

```python
popular_tracks
```

```python
#треки прослушанные пользователями
listening_by_users = popular_tracks[popular_tracks['track_id'].isin(listening['track_id'])]
```

```python
listening_by_users
```

```python
# столько прослушали пользователи треков из 5 популярных жанров 
filtered_listening = listening.drop('playcount', axis=1)[listening['track_id'].isin(listening_by_users['track_id'])]
filtered_listening
```

```python
most_popular = listening_by_users[['genre', 'track_id']].merge(filtered_listening)
most_popular
```

```python
#df = most_popular.groupby(['genre', 'track_id']).count().sort_values('user_id').sort_values('genre').groupby('genre').head(5)
#most_popular.sort_values('user_id', ascending=False).groupby(['genre', 'track_id']).count()
popular_by_genres = most_popular.groupby(['genre', 'track_id']).count().sort_values('user_id', ascending=False)
popular_by_genres
```

```python
most_listened_tracks = popular_by_genres.groupby('genre').head(5).reset_index()['track_id']
most_listened_tracks

```

```python
listening_most_popular_tracks = listening.drop('playcount', axis=1)[listening['track_id'].isin(most_listened_tracks)]
listening_most_popular_tracks
```

```python
listened_by_users = listening_most_popular_tracks.groupby('user_id').count().sort_values('track_id')
listened_by_users[listened_by_users['track_id'] > 5]
```

```python
group_1 = listened_by_users.tail(2)

group_2 = listened_by_users[listened_by_users['track_id'] == 15].head(2)

group_3 = listened_by_users[listened_by_users['track_id'] == 12].head(2)

group_4 = listened_by_users[listened_by_users['track_id'] == 9].head(2)

group_5 = listened_by_users[listened_by_users['track_id'] == 6].head(2)
```

```python
users = pd.concat([group_1, group_2, group_3, group_4, group_5]).reset_index()['user_id']
users
df_users =  pd.DataFrame(users)
```

```python
df_tracks = data[['track_id', 'name', 'artist', 'genre']][data['track_id'].isin(most_listened_tracks)].reset_index(drop=True)
df_tracks
```

```python
most_listened_tracks
```

```python
df_listening = listening_most_popular_tracks[listening_most_popular_tracks['track_id'].isin(most_listened_tracks)][listening_most_popular_tracks['user_id'].isin(users)].reset_index(drop=True)
df_listening
```
